```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset_study.csv')
df.head()
```

```python
df.shape
```

```python
X=df['study_hours'].values.reshape(-1,1)
```

```python
Y=df['grade'].values.reshape(-1,1)
```

```python
X_train = X[:20]
Y_train = Y[:20]
```

```python
X_train = X_train.reshape(-1,1)
Y_train = Y_train.reshape(-1,1)
```

```python
plt.scatter(df['study_hours'], df['grade'])
plt.xlabel('study_hours')
plt.ylabel('grade')
plt.title('study_hours v/s grade')

X = df['study_hours'].values
plt.plot(X, 8* X + 20)
plt.show()
```

```python
class Linear_Regression():

    # constructor to the class,provoding values
    def __init__(self, learning_rate, no_of_itr):
        self.learning_rate = learning_rate
        self.no_of_itr = no_of_itr
        
    def fit(self, X, Y):

        # No. of training examples and no. of features.
        self.m, self.n = X.shape     # Number of rows and columns
        # Initiating the weight and bias
        self.w = np.zeros((self.n, 1))
        self.b = 0
        self.X = X
        self.Y = Y

        # Implementing the gradient descent.
        for i in range(self.no_of_itr):
            self.update_weigths()

    def update_weigths(self):
        Y_prediction = self.predict(self.X)

        # Calculating gradients
        dw = -(self.X.T).dot(self.Y - Y_prediction)/self.m

        db = -np.sum(self.Y - Y_prediction)/self.m

        # Updating weights
        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db

    def predict(self, X):
        return X.dot(self.w) + self.b

    def print_weights(self):
        print('Weights for the respective features are :')
        print(self.w)
        print()

        print('Bias value for the regression is ', self.b)
```

```python
model = Linear_Regression(learning_rate=0.03,
                          no_of_itr=2000)
model.fit(X_train, Y_train)
```

```python
model.print_weights()
```

```python
plt.scatter(df['study_hours'], df['grade'])

X = df['study_hours'].values.reshape(-1,1)

y_pred = model.predict(X)

plt.plot(df['study_hours'], y_pred)

plt.show()
```
